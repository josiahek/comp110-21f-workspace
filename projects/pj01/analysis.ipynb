## Steps needed to test hypothesis:

3. Count how many students are in each graduating class

## Step 1: Reading the csv file & creating a column-based table:

```python
from data_utils import read_csv_rows, columnar, head
from tabulate import tabulate

data_file: str = "../../data/survey.csv"
data_rows: list[dict[str, str]] = read_csv_rows(data_file)
col_table: dict[str, list[str]] = columnar(data_rows)

first_rows: dict[str, list[str]] = head(col_table, 5)
tabulate(first_rows, first_rows.keys(), "html")
```

## Step 2: Selecting the values in the year column:

```python
from data_utils import select

year_subset: dict[str, list[str]] = select(col_table, ["year"])
year_subset
```

## Step 3: Counting the total of undergrads in each graduating class:

```python
from data_utils import count

counter = count(year_subset['year'])

print(f"Year 22 & 21 (Senior) total is: {counter['22'] + counter['21']}")
print(f"Year 23 (Junior) total is: {counter['23']}")
print(f"Year 24 (Sophomore) total is: {counter['24']}")
print(f"Year 25 (First-year) total is: {counter['25']}")
```

## Step 4: Graphing the results
